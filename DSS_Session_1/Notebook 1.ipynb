## Using cx_Oracle and pandas to inspect the Archive

### Querying for all the available Tables

We will start by loading the libraries, read a *connection string* (to be discussed live in our next session), and open a connection to the read-only version of the archive.

````python
# Any line beggining in a code-cel with a # is a comment line, and won't be 
# executed

import os
import pandas as pd
import cx_Oracle

connection_string = os.environ['CON_STR']
````

````python
# Create connection to the read-only DB
con = cx_Oracle.connect(connection_string)
````

````python
# Create a cursor to do SQL queries. The cursor is the object that is used to do 
# queries to the archive
cur = con.cursor()
````

To execute an SQL query we use the `execute` method of the `cur` object we just created (`cur.execute()`). The SQL queries are strings. For example, let's query for all the available tables in the archive. The query string we will store in `sql_all_tables`

````python
sql_all_tables = """
SELECT DISTINCT OWNER, OBJECT_NAME, OBJECT_TYPE
  FROM ALL_OBJECTS
 WHERE OBJECT_TYPE = 'TABLE'
"""
````

Next, we request to the database to execute the query:

````python
cur.execute(sql_all_tables)
````

And to get the results we use the method `fetchall`:

````python
cur.fetchall()
````

`fetchall` returns the query results in a list, each row represented by a tuple, with one element by each column queried. It is important to notice that once we have requested the results from the query, execution `fetchall` once more will produce not results:

````python
cur.fetchall()
````

So, to get again the results one needs to run `cur.execute` again, followed by the `cur.fetchall` call.

We have not stored the results in any local variable yet. Let's do that. But since we want to start using pandas, we will store the results in a pandas Dataframe, which is one of the two data structures provided by `pandas`:

![img](img/pandas-02.png)

(see [pandas data structures](http://pandas.pydata.org/pandas-docs/stable/overview.html#data-structures) and [pandas documentation](http://pandas.pydata.org/pandas-docs/stable/index.html) in general for a full description of a Dataframe).

How do we create a dataframe to store the queries' results?

````python
# execute query again
cur.execute(sql_all_tables)
# fetch the results and use them to create the dataframe
all_tables_df = pd.DataFrame(
    cur.fetchall(), columns=[rec[0] for rec in cur.description])
````

Creating a pandas DataFrame is done by calling `pd.Dataframes`. A dataframe can be created with multiple kinds of iterables (lists, tuples, sets, numpy arrays, etc.): in this case we are using the list of tuples that is returned by `cur.fetchall`. The second named argument, `columns=[some_list]`, is used to set the names for the columns of the Dataframe. T

In this case the names of the columns were created using the [list comprehension](https://docs.python.org/2/tutorial/datastructures.html#list-comprehensions) `[rec[0] for rec in cur.description]`. It is left as an exercise workout what exactly is going on here, but **this is a method that you can use every time you create a dataframe from a cx_oracle query**.

## Shiftlog Database

We will show here how the shiftlog data is stored in the archive. There are 12 tables that are populated by the Shiftlog subsystem in the ALMA Schema, of which 4 will be relevant for our work:

1. ALMA.SHIFTLOG_ENTRIES
2. ALMA.SLOG_ENTRY_ATTR
3. ALMA.SLOG_ALLOCATED_TIME_INTERVAL
4. ALMA.SHIFTLOG_REPLY

For our exercise, let's query a subsample from these tables, with all the data since January 1st, 2018:

````python
# Define our SQL queries:

sql_shiftlog_entries = """
    SELECT * 
      FROM ALMA.SHIFTLOG_ENTRIES 
     WHERE SE_TIMESTAMP > TO_TIMESTAMP(:datestart, 'YYYY-MM-DD HH24:MI:SS')
"""

sql_slog_allocated = """
    SELECT *
      FROM ALMA.SLOG_ALLOCATED_TIME_INTERVAL
     WHERE STARTTIME > TO_TIMESTAMP(:datestart, 'YYYY-MM-DD HH24:MI:SS')
"""

# The following two queries are one step more advanced, because we query
# the data based in a restriction comming from another table. More in the next
# live tutorial session.

sql_slog_entry_attr = """
    SELECT sea.* 
      FROM ALMA.SHIFTLOG_ENTRIES sa, ALMA.SLOG_ENTRY_ATTR sea
     WHERE sa.SE_TIMESTAMP > TO_TIMESTAMP(:datestart, 'YYYY-MM-DD HH24:MI:SS')
      AND sa.SE_ID = sea.SLOG_SE_ID
"""

sql_shiftlog_reply = """
    SELECT sr.* 
      FROM ALMA.SHIFTLOG_ENTRIES se, ALMA.SHIFTLOG_REPLY sr
     WHERE se.SE_TIMESTAMP > TO_TIMESTAMP(:datestart, 'YYYY-MM-DD HH24:MI:SS')
      AND se.SE_ID = sr.ENTRY_ID
"""
````

You might have noticed in the previous queries the use of *binding variables*, which are the variables that begin with a colon, in this case `:datestart`. When we do this, we make our queries more efficient, but the technical reasons are beyond the scope of this notebook (and from the knowledge of the author, so we might invite an APO member to give us more background if needed). 

The binding variables are used with the method `.excute`, and here are the examples:

````python
# Create shiftlog_entries_df dataframe for the data 
# from ALMA.SHIFTLOG_ENTRIES
cur.execute(sql_shiftlog_entries, datestart='2018-01-01 00:00:00')
shiftlog_entries_df = pd.DataFrame(
    cur.fetchall(), columns=[rec[0] for rec in cur.description])

# Create allocated_time_df dataframe for ALMA.SLOG_ALLOCATED_TIME_INTERVAL
# data
cur.execute(sql_slog_allocated, datestart='2018-01-01 00:00:00')
allocated_time_df = pd.DataFrame(
    cur.fetchall(), columns=[rec[0] for rec in cur.description])

# Create the entries_attributes_df dataframe for ALMA.SLOG_ENTRY_ATTR
cur.execute(sql_slog_entry_attr, datestart='2018-01-01 00:00:00')
entries_attributes_df = pd.DataFrame(
    cur.fetchall(), columns=[rec[0] for rec in cur.description])

# Create the shiftlog_comments_df dataframe for ALMA.SHIFTLOG_REPLY
cur.execute(sql_shiftlog_reply, datestart='2018-01-01 00:00:00')
shiftlog_comments_df = pd.DataFrame(
    cur.fetchall(), columns=[rec[0] for rec in cur.description])
````

So, the shiflog tool helps to visually inspect the entries that match some criteria that can be provided through the Query Parameters interface in the bottom right box. This tell us that we have 11 different types or categories for the entries, and from the wiki we find that the categories are:

* *Shift entry (SHIFT):* should be created every time a shift starts.
* *Manual mode execution entries* *(MMEX):* These are created automatically every time a script is executed on a manual array.
* *Scheduling block execution entries (SBEX):* whenever a SB is executed on an interactive array, a SBEX entry is created with all info about the project.
* *Antenna entries (ANTEN):* automatically created every time the "antenna" button (top-right) is pressed (and at the beginning of every shift) to store information about the currently configured antennas on the system.
* *Weather entries (WEATH):* automatically created every 15 min. or every time the "weather" button (top right) is pressed.
* *Array entries (ARRAY):* automatically created to store information about the arrays that are created in the system.
* *Downtime entries (DOWN):* for including manually information on downtime due to technical problems or bad weather (icon at top-left).
* *General entries (ENTRY):* this one should be used for manually created entries (icon at top-left).
* *Engineering entries (ENGIN)*: to take note of the main activities during the engineering time (icon at top-left).
* *Command line entries *(CMDLN):* for recording scripts that do not produce and asdm and that are related to control and operational processes during the observations (e.g. for !RadioSetup.py)
* *Array Operations Group entries(AOG):* for allowing operators to communicate information

Each kind of entry can be accessed by queries, once you know their type's ID (`SE_TYPE`) within the Database.

#### Shift Entries: SE_TYPE 1

The shift entries are of SE_TYPE = 1. We will use the dataframe methods to do a query (`.query`) and to remove the columns that are not relevant for each entry (`.dropna`)

````python
shift_entries = shiftlog_entries_df.query('SE_TYPE == 1').dropna(axis=1, how='all').copy()
````

````python
shift_entries.info()
````

````python
shift_entries.head()
````

The relevant columns, by their meaning:

* SE_ID: unique ID for a shiftlog entry
* SE_TYPE: type or category for an entry (in this case all 1 by construction)
* SE_SUBJECT: comment written by the system or operator when creating the shift
* SE_TIMESTAMP: end time for the shit entry. Is equal to SE_START when the shift is ongoing.
* SE_AUTHOR: user who created the entry
* SE_START: start time for the entry
* SE_LOCATION: STE originating the shiftlog entry
* SE_ALMABUILD: ALMA software version active **at the beginning of the shift** (if there is a change of software after the shift is started, this information won't be reflected here)
* SE_ACSVERSION: ACS version active at the beginning of the shift.
* SE_SHIFTACTIVITY: Main shift activity, can be any of {{shift_entries.SE_SHIFTACTIVITY.unique().tolist()}} (and maybe others)
* SE_PWV: PWV, not to be used as it doesn't have a clear meaning for a shift, and sometimes is even empty.

How long are the Shifts? Let's create a new column with the time elapsed, in hours, between the start of the shift (SE_START) and the end (SE_TIMESTAMP) (Shifts that are ongoing have the same time in both columns):

````python
shift_entries['TimeElapsed'] = (shift_entries.SE_TIMESTAMP - shift_entries.SE_START).apply(
    lambda x: x.total_seconds() / 3600.)
````

**Note about the code in previous cell**: The output of `shift_entries.SE_TIMESTAMP - shift_entries.SE_START` is a pandas Series, and this is why we can use `.apply` to run a function over this output. In this case the Series contains a `pandas.Timedelta` data type, which has a method `total_seconds` available, that transforms the time delta into seconds. By dividing by 3600.0 we get back the hours. We will talk more about .apply in the live sessions.

Now, let's inspect the first 10 rows, and only display the most significant columns of our table, with the new result:

````python
shift_entries.head(10)[['SE_ID', 'SE_LOCATION', 'SE_START', 'SE_TIMESTAMP','TimeElapsed', 'SE_SHIFTACTIVITY']]
````

We could now look, for example, for Shifts entries that might be wrong! And this is a good exercise when using data from any database: we should never assume the data is clean and ready to be ingested! For example, any shift lasting more than 32 hours might be a wrong entry, or any shift with negative time:

````python
shift_entries.query('TimeElapsed > 32 or TimeElapsed < 0')[
    ['SE_ID', 'SE_LOCATION', 'SE_START', 'SE_TIMESTAMP','TimeElapsed', 'SE_SHIFTACTIVITY']]
````

Now we could ask for all the valid science Shifts since January, ordered by descending Start Time:

````python
# Notice the format for the python code: this is being use a lot in data science
# when one is applying one pandas method over the result of another:

shift_entries.query(
    'SE_SHIFTACTIVITY == "SciOps"'
).sort_values(
    by='SE_START', ascending=False
)[
    ['SE_ID', 'SE_LOCATION', 'SE_START', 'SE_TIMESTAMP','TimeElapsed', 
     'SE_SHIFTACTIVITY']
]
````
